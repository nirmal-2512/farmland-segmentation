```python
import os
import cv2
import numpy as np

MASK_DIR = r"D:\farmland boundary\dataset\masks"

BOUNDARY_DIR = r"D:\farmland boundary\dataset\boundary_masks"

os.makedirs(
    BOUNDARY_DIR,
    exist_ok=True
)

kernel = np.ones((3,3),np.uint8)

for file in os.listdir(MASK_DIR):

    mask_path = os.path.join(
        MASK_DIR,
        file
    )

    mask = cv2.imread(
        mask_path,
        0
    )

    _, mask = cv2.threshold(
        mask,
        127,
        255,
        cv2.THRESH_BINARY
    )

    eroded = cv2.erode(
        mask,
        kernel,
        iterations=1
    )

    boundary = mask - eroded

    save_path = os.path.join(
        BOUNDARY_DIR,
        file
    )

    cv2.imwrite(
        save_path,
        boundary
    )

print("Boundary masks created")
print(
    "Count:",
    len(os.listdir(BOUNDARY_DIR))
)
```

```python
import os
import cv2
import numpy as np
from pathlib import Path

import torch
import torch.nn as nn
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

import albumentations as A
from albumentations.pytorch import ToTensorV2

from sklearn.model_selection import train_test_split

import segmentation_models_pytorch as smp

# =====================================================
# CONFIG
# =====================================================

IMAGE_SIZE = 768

BATCH_SIZE = 2

EPOCHS = 80

LR = 1e-4

DEVICE = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)

IMAGE_DIR = r"D:\farmland boundary\dataset\images"

MASK_DIR = r"D:\farmland boundary\dataset\masks"

BOUNDARY_DIR = r"D:\farmland boundary\dataset\boundary_masks"

MODEL_NAME = (
    "farmland_boundary_dualhead_b3.pth"
)

# =====================================================
# FILE LIST
# =====================================================

image_files = sorted([
    f
    for f in os.listdir(IMAGE_DIR)
    if f.lower().endswith(
        (".jpg",".jpeg",".png")
    )
])

train_files,val_files = train_test_split(
    image_files,
    test_size=0.2,
    random_state=42
)

# =====================================================
# AUGMENTATION
# =====================================================

train_tf = A.Compose([

    A.Resize(
        IMAGE_SIZE,
        IMAGE_SIZE
    ),

    A.HorizontalFlip(p=0.5),

    A.VerticalFlip(p=0.5),

    A.RandomRotate90(p=0.5),

    A.RandomBrightnessContrast(p=0.5),

    ToTensorV2()

])

val_tf = A.Compose([

    A.Resize(
        IMAGE_SIZE,
        IMAGE_SIZE
    ),

    ToTensorV2()

])

# =====================================================
# DATASET
# =====================================================

class FarmDataset(Dataset):

    def __init__(
        self,
        files,
        transform
    ):

        self.files = files

        self.transform = transform

    def __len__(self):

        return len(self.files)

    def __getitem__(
        self,
        idx
    ):

        file = self.files[idx]

        image = cv2.imread(
            os.path.join(
                IMAGE_DIR,
                file
            )
        )

        image = cv2.cvtColor(
            image,
            cv2.COLOR_BGR2RGB
        )

        mask_name = (
            Path(file).stem
            + ".png"
        )

        farmland = cv2.imread(
            os.path.join(
                MASK_DIR,
                mask_name
            ),
            0
        )

        boundary = cv2.imread(
            os.path.join(
                BOUNDARY_DIR,
                mask_name
            ),
            0
        )

        farmland = (
            farmland > 127
        ).astype(np.float32)

        boundary = (
            boundary > 127
        ).astype(np.float32)

        transformed = self.transform(

            image=image,

            masks=[
                farmland,
                boundary
            ]
        )

        image = (
            transformed["image"]
            .float()
            /255.0
        )

        farmland = (
            transformed["masks"][0]
            .unsqueeze(0)
        )

        boundary = (
            transformed["masks"][1]
            .unsqueeze(0)
        )

        return (
            image,
            farmland,
            boundary
        )

# =====================================================
# MODEL
# =====================================================

class DualHeadUNetPP(
    nn.Module
):

    def __init__(self):

        super().__init__()

        self.backbone = smp.UnetPlusPlus(

            encoder_name=
            "efficientnet-b3",

            encoder_weights=
            "imagenet",

            in_channels=3,

            classes=32
        )

        self.farmland_head = nn.Conv2d(
            32,
            1,
            1
        )

        self.boundary_head = nn.Conv2d(
            32,
            1,
            1
        )

    def forward(
        self,
        x
    ):

        feat = self.backbone(x)

        farmland = (
            self.farmland_head(feat)
        )

        boundary = (
            self.boundary_head(feat)
        )

        return (
            farmland,
            boundary
        )

# =====================================================
# LOADERS
# =====================================================

train_ds = FarmDataset(
    train_files,
    train_tf
)

val_ds = FarmDataset(
    val_files,
    val_tf
)

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH_SIZE,
    shuffle=True
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH_SIZE
)

# =====================================================
# MODEL
# =====================================================

model = (
    DualHeadUNetPP()
    .to(DEVICE)
)

dice = smp.losses.DiceLoss(
    mode="binary"
)

bce = nn.BCEWithLogitsLoss()

# =====================================================
# LOSS
# =====================================================

def total_loss(

    pred_farmland,

    pred_boundary,

    gt_farmland,

    gt_boundary

):

    farmland_loss = (

        0.7*dice(
            pred_farmland,
            gt_farmland
        )

        +

        0.3*bce(
            pred_farmland,
            gt_farmland
        )
    )

    boundary_loss = (

        0.7*dice(
            pred_boundary,
            gt_boundary
        )

        +

        0.3*bce(
            pred_boundary,
            gt_boundary
        )
    )

    return (

        0.7*farmland_loss

        +

        0.3*boundary_loss

    )

# =====================================================
# OPTIMIZER
# =====================================================

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=LR
)

best_val = 999

# =====================================================
# TRAIN
# =====================================================

for epoch in range(EPOCHS):

    model.train()

    train_loss = 0

    for image, farmland, boundary in train_loader:

        image = image.to(DEVICE)

        farmland = farmland.to(
            DEVICE
        )

        boundary = boundary.to(
            DEVICE
        )

        optimizer.zero_grad()

        pred_farmland,pred_boundary = model(image)

        loss = total_loss(

            pred_farmland,

            pred_boundary,

            farmland,

            boundary
        )

        loss.backward()

        optimizer.step()

        train_loss += loss.item()

    train_loss /= len(
        train_loader
    )

    print(
        f"Epoch {epoch+1}"
        f" Loss {train_loss:.4f}"
    )

torch.save(
    model.state_dict(),
    MODEL_NAME
)

print(
    "Training Finished"
)
```
