```python
import os
import cv2
import numpy as np
import xml.etree.ElementTree as ET
from pathlib import Path

# ==========================================================
# OUTPUT FOLDER
# ==========================================================

OUTPUT_MASK_DIR = r"D:\farmland boundary\dataset\masks"

os.makedirs(OUTPUT_MASK_DIR, exist_ok=True)

# ==========================================================
# XML FILES
# ==========================================================

xml_files = [
    r"D:\farmland boundary\New folder\Train\boundary mask\Train1_annotations.xml",
    r"D:\farmland boundary\New folder\Train 2\boundarydet\Train2_annotations.xml"
]

# ==========================================================
# PROCESS BOTH XML FILES
# ==========================================================

for xml_path in xml_files:

    print("\n===================================")
    print("Processing XML:")
    print(xml_path)
    print("===================================\n")

    tree = ET.parse(xml_path)
    root = tree.getroot()

    image_count = 0

    for image_tag in root.findall("image"):

        image_name = image_tag.attrib["name"]

        width = int(image_tag.attrib["width"])
        height = int(image_tag.attrib["height"])

        mask = np.zeros((height, width), dtype=np.uint8)

        # ------------------------------------
        # ONLY FARMLAND POLYGONS
        # ------------------------------------

        for polygon in image_tag.findall("polygon"):

            label = polygon.attrib["label"]

            if label.lower() != "farmland":
                continue

            points = []

            for point in polygon.attrib["points"].split(";"):

                x, y = map(float, point.split(","))

                points.append([
                    int(round(x)),
                    int(round(y))
                ])

            points = np.array([points], dtype=np.int32)

            cv2.fillPoly(mask, points, 255)

        # ------------------------------------
        # SAVE MASK
        # ------------------------------------

        mask_name = Path(image_name).stem + ".png"

        mask_path = os.path.join(
            OUTPUT_MASK_DIR,
            mask_name
        )

        cv2.imwrite(mask_path, mask)

        image_count += 1

    print(f"Saved {image_count} masks")

print("\nDONE")
```

```python
import os
import cv2
import numpy as np
from pathlib import Path

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

import albumentations as A
from albumentations.pytorch import ToTensorV2

import segmentation_models_pytorch as smp

from sklearn.model_selection import train_test_split

# =====================================================
# CONFIG
# =====================================================

IMAGE_SIZE = 768

BATCH_SIZE = 2

EPOCHS = 100

LR = 1e-4

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

IMAGE_DIR = r"D:\farmland boundary\dataset\images"

MASK_DIR = r"D:\farmland boundary\dataset\masks"

MODEL_NAME = "farmland_boundary_unetpp_b3.pth"

# =====================================================
# FILES
# =====================================================

image_files = sorted([
    x for x in os.listdir(IMAGE_DIR)
    if x.lower().endswith(
        (".jpg",".jpeg",".png")
    )
])

train_files, val_files = train_test_split(
    image_files,
    test_size=0.2,
    random_state=42
)

# =====================================================
# AUGMENTATION
# =====================================================

train_transform = A.Compose([

    A.Resize(
        IMAGE_SIZE,
        IMAGE_SIZE
    ),

    A.HorizontalFlip(p=0.5),

    A.VerticalFlip(p=0.5),

    A.RandomRotate90(p=0.5),

    A.RandomBrightnessContrast(p=0.5),

    A.HueSaturationValue(p=0.3),

    A.GaussNoise(p=0.2),

    ToTensorV2()

])

val_transform = A.Compose([

    A.Resize(
        IMAGE_SIZE,
        IMAGE_SIZE
    ),

    ToTensorV2()

])

# =====================================================
# DATASET
# =====================================================

class FarmlandDataset(Dataset):

    def __init__(
        self,
        files,
        transform=None
    ):

        self.files = files

        self.transform = transform

    def __len__(self):

        return len(self.files)

    def __getitem__(self, idx):

        img_name = self.files[idx]

        image_path = os.path.join(
            IMAGE_DIR,
            img_name
        )

        mask_name = (
            Path(img_name).stem
            + ".png"
        )

        mask_path = os.path.join(
            MASK_DIR,
            mask_name
        )

        image = cv2.imread(image_path)

        image = cv2.cvtColor(
            image,
            cv2.COLOR_BGR2RGB
        )

        mask = cv2.imread(
            mask_path,
            0
        )

        mask = (
            mask > 127
        ).astype(np.float32)

        transformed = self.transform(
            image=image,
            mask=mask
        )

        image = transformed["image"].float()/255.0

        mask = transformed["mask"]

        mask = mask.unsqueeze(0)

        return image, mask

# =====================================================
# LOADERS
# =====================================================

train_ds = FarmlandDataset(
    train_files,
    train_transform
)

val_ds = FarmlandDataset(
    val_files,
    val_transform
)

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=0
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=0
)

# =====================================================
# MODEL
# =====================================================

model = smp.UnetPlusPlus(

    encoder_name="efficientnet-b3",

    encoder_weights="imagenet",

    in_channels=3,

    classes=1

)

model = model.to(DEVICE)

# =====================================================
# LOSS
# =====================================================

dice_loss = smp.losses.DiceLoss(
    mode="binary"
)

bce_loss = nn.BCEWithLogitsLoss()

def loss_fn(pred, target):

    return (

        0.7 * dice_loss(
            pred,
            target
        )

        +

        0.3 * bce_loss(
            pred,
            target
        )

    )

# =====================================================
# OPTIMIZER
# =====================================================

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=LR
)

scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(
    optimizer,
    T_max=EPOCHS
)

scaler = torch.cuda.amp.GradScaler()

best_loss = 999999

# =====================================================
# TRAINING
# =====================================================

for epoch in range(EPOCHS):

    model.train()

    train_loss = 0

    for images,masks in train_loader:

        images = images.to(DEVICE)

        masks = masks.to(DEVICE)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast():

            outputs = model(images)

            loss = loss_fn(
                outputs,
                masks
            )

        scaler.scale(loss).backward()

        scaler.step(optimizer)

        scaler.update()

        train_loss += loss.item()

    scheduler.step()

    # -------------------

    model.eval()

    val_loss = 0

    with torch.no_grad():

        for images,masks in val_loader:

            images = images.to(DEVICE)

            masks = masks.to(DEVICE)

            outputs = model(images)

            loss = loss_fn(
                outputs,
                masks
            )

            val_loss += loss.item()

    train_loss /= len(train_loader)

    val_loss /= len(val_loader)

    print(
        f"Epoch {epoch+1}/{EPOCHS}"
        f" | Train {train_loss:.4f}"
        f" | Val {val_loss:.4f}"
    )

    if val_loss < best_loss:

        best_loss = val_loss

        torch.save(
            model.state_dict(),
            MODEL_NAME
        )

        print(
            "Saved Best Model"
        )

print("Training Finished")
```

```python
import cv2
import torch
import numpy as np
import segmentation_models_pytorch as smp

DEVICE = "cuda"

model = smp.UnetPlusPlus(
    encoder_name="efficientnet-b3",
    encoder_weights=None,
    in_channels=3,
    classes=1
)

model.load_state_dict(
    torch.load(
        "farmland_boundary_unetpp_b3.pth",
        map_location=DEVICE
    )
)

model.eval()
model.to(DEVICE)

img = cv2.imread(
    r"D:\farmland boundary\dataset\images\102.jpg"
)

img_rgb = cv2.cvtColor(
    img,
    cv2.COLOR_BGR2RGB
)

img_rgb = cv2.resize(
    img_rgb,
    (768,768)
)

x = (
    torch.tensor(img_rgb)
    .permute(2,0,1)
    .float()/255.0
)

x = x.unsqueeze(0).to(DEVICE)

with torch.no_grad():

    pred = model(x)

pred = torch.sigmoid(pred)

pred = pred.squeeze().cpu().numpy()

mask = (pred > 0.5).astype(np.uint8)*255

cv2.imwrite(
    "test_prediction.png",
    mask
)
```
