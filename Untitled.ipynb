```python
import os
import cv2
import numpy as np

MASK_DIR = r"D:\farmland boundary\dataset\masks"

BOUNDARY_DIR = r"D:\farmland boundary\dataset\boundary_masks"

os.makedirs(
    BOUNDARY_DIR,
    exist_ok=True
)

kernel = np.ones((3,3),np.uint8)

for file in os.listdir(MASK_DIR):

    mask_path = os.path.join(
        MASK_DIR,
        file
    )

    mask = cv2.imread(
        mask_path,
        0
    )

    _, mask = cv2.threshold(
        mask,
        127,
        255,
        cv2.THRESH_BINARY
    )

    eroded = cv2.erode(
        mask,
        kernel,
        iterations=1
    )

    boundary = mask - eroded

    save_path = os.path.join(
        BOUNDARY_DIR,
        file
    )

    cv2.imwrite(
        save_path,
        boundary
    )

print("Boundary masks created")
print(
    "Count:",
    len(os.listdir(BOUNDARY_DIR))
)
```

```python
import os
import cv2
import numpy as np
from pathlib import Path

import torch
import torch.nn as nn
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

import albumentations as A
from albumentations.pytorch import ToTensorV2

from sklearn.model_selection import train_test_split

import segmentation_models_pytorch as smp

# =====================================================
# CONFIG
# =====================================================

IMAGE_SIZE = 768

BATCH_SIZE = 2

EPOCHS = 80

LR = 1e-4

DEVICE = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)

IMAGE_DIR = r"D:\farmland boundary\dataset\images"

MASK_DIR = r"D:\farmland boundary\dataset\masks"

BOUNDARY_DIR = r"D:\farmland boundary\dataset\boundary_masks"

MODEL_NAME = (
    "farmland_boundary_dualhead_b3.pth"
)

# =====================================================
# FILE LIST
# =====================================================

image_files = sorted([
    f
    for f in os.listdir(IMAGE_DIR)
    if f.lower().endswith(
        (".jpg",".jpeg",".png")
    )
])

train_files,val_files = train_test_split(
    image_files,
    test_size=0.2,
    random_state=42
)

# =====================================================
# AUGMENTATION
# =====================================================

train_tf = A.Compose([

    A.Resize(
        IMAGE_SIZE,
        IMAGE_SIZE
    ),

    A.HorizontalFlip(p=0.5),

    A.VerticalFlip(p=0.5),

    A.RandomRotate90(p=0.5),

    A.RandomBrightnessContrast(p=0.5),

    ToTensorV2()

])

val_tf = A.Compose([

    A.Resize(
        IMAGE_SIZE,
        IMAGE_SIZE
    ),

    ToTensorV2()

])

# =====================================================
# DATASET
# =====================================================

class FarmDataset(Dataset):

    def __init__(
        self,
        files,
        transform
    ):

        self.files = files

        self.transform = transform

    def __len__(self):

        return len(self.files)

    def __getitem__(
        self,
        idx
    ):

        file = self.files[idx]

        image = cv2.imread(
            os.path.join(
                IMAGE_DIR,
                file
            )
        )

        image = cv2.cvtColor(
            image,
            cv2.COLOR_BGR2RGB
        )

        mask_name = (
            Path(file).stem
            + ".png"
        )

        farmland = cv2.imread(
            os.path.join(
                MASK_DIR,
                mask_name
            ),
            0
        )

        boundary = cv2.imread(
            os.path.join(
                BOUNDARY_DIR,
                mask_name
            ),
            0
        )

        farmland = (
            farmland > 127
        ).astype(np.float32)

        boundary = (
            boundary > 127
        ).astype(np.float32)

        transformed = self.transform(

            image=image,

            masks=[
                farmland,
                boundary
            ]
        )

        image = (
            transformed["image"]
            .float()
            /255.0
        )

        farmland = (
            transformed["masks"][0]
            .unsqueeze(0)
        )

        boundary = (
            transformed["masks"][1]
            .unsqueeze(0)
        )

        return (
            image,
            farmland,
            boundary
        )

# =====================================================
# MODEL
# =====================================================

class DualHeadUNetPP(
    nn.Module
):

    def __init__(self):

        super().__init__()

        self.backbone = smp.UnetPlusPlus(

            encoder_name=
            "efficientnet-b3",

            encoder_weights=
            "imagenet",

            in_channels=3,

            classes=32
        )

        self.farmland_head = nn.Conv2d(
            32,
            1,
            1
        )

        self.boundary_head = nn.Conv2d(
            32,
            1,
            1
        )

    def forward(
        self,
        x
    ):

        feat = self.backbone(x)

        farmland = (
            self.farmland_head(feat)
        )

        boundary = (
            self.boundary_head(feat)
        )

        return (
            farmland,
            boundary
        )

# =====================================================
# LOADERS
# =====================================================

train_ds = FarmDataset(
    train_files,
    train_tf
)

val_ds = FarmDataset(
    val_files,
    val_tf
)

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH_SIZE,
    shuffle=True
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH_SIZE
)

# =====================================================
# MODEL
# =====================================================

model = (
    DualHeadUNetPP()
    .to(DEVICE)
)

dice = smp.losses.DiceLoss(
    mode="binary"
)

bce = nn.BCEWithLogitsLoss()

# =====================================================
# LOSS
# =====================================================

def total_loss(

    pred_farmland,

    pred_boundary,

    gt_farmland,

    gt_boundary

):

    farmland_loss = (

        0.7*dice(
            pred_farmland,
            gt_farmland
        )

        +

        0.3*bce(
            pred_farmland,
            gt_farmland
        )
    )

    boundary_loss = (

        0.7*dice(
            pred_boundary,
            gt_boundary
        )

        +

        0.3*bce(
            pred_boundary,
            gt_boundary
        )
    )

    return (

        0.7*farmland_loss

        +

        0.3*boundary_loss

    )

# =====================================================
# OPTIMIZER
# =====================================================

optimizer = torch.optim.AdamW(
    model.parameters(),
    lr=LR
)

best_val = 999

# =====================================================
# TRAIN
# =====================================================

for epoch in range(EPOCHS):

    model.train()

    train_loss = 0

    for image, farmland, boundary in train_loader:

        image = image.to(DEVICE)

        farmland = farmland.to(
            DEVICE
        )

        boundary = boundary.to(
            DEVICE
        )

        optimizer.zero_grad()

        pred_farmland,pred_boundary = model(image)

        loss = total_loss(

            pred_farmland,

            pred_boundary,

            farmland,

            boundary
        )

        loss.backward()

        optimizer.step()

        train_loss += loss.item()

    train_loss /= len(
        train_loader
    )

    print(
        f"Epoch {epoch+1}"
        f" Loss {train_loss:.4f}"
    )

torch.save(
    model.state_dict(),
    MODEL_NAME
)

print(
    "Training Finished"
)
```

```python
import os
import cv2

IMG_DIR = r"D:\farmland boundary\dataset\images"
MASK_DIR = r"D:\farmland boundary\dataset\masks"

PATCH_IMG_DIR = r"D:\farmland boundary\patches\images"
PATCH_MASK_DIR = r"D:\farmland boundary\patches\masks"

os.makedirs(PATCH_IMG_DIR, exist_ok=True)
os.makedirs(PATCH_MASK_DIR, exist_ok=True)

PATCH_SIZE = 768
STRIDE = 384

count = 0

for file in os.listdir(IMG_DIR):

    img_path = os.path.join(IMG_DIR, file)

    mask_path = os.path.join(
        MASK_DIR,
        os.path.splitext(file)[0] + ".png"
    )

    img = cv2.imread(img_path)
    mask = cv2.imread(mask_path, 0)

    if img is None or mask is None:
        continue

    H, W = img.shape[:2]

    # Ensure rightmost and bottom areas are included
    xs = list(range(0, max(W - PATCH_SIZE + 1, 1), STRIDE))
    ys = list(range(0, max(H - PATCH_SIZE + 1, 1), STRIDE))

    if xs[-1] != W - PATCH_SIZE:
        xs.append(W - PATCH_SIZE)

    if ys[-1] != H - PATCH_SIZE:
        ys.append(H - PATCH_SIZE)

    for y in ys:
        for x in xs:

            img_patch = img[
                y:y+PATCH_SIZE,
                x:x+PATCH_SIZE
            ]

            mask_patch = mask[
                y:y+PATCH_SIZE,
                x:x+PATCH_SIZE
            ]

            if img_patch.shape[:2] != (PATCH_SIZE, PATCH_SIZE):
                continue

            cv2.imwrite(
                os.path.join(
                    PATCH_IMG_DIR,
                    f"{count}.png"
                ),
                img_patch
            )

            cv2.imwrite(
                os.path.join(
                    PATCH_MASK_DIR,
                    f"{count}.png"
                ),
                mask_patch
            )

            count += 1

print("Total patches:", count)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import segmentation_models_pytorch as smp

dice_loss = smp.losses.DiceLoss(
    mode="binary"
)

bce_loss = nn.BCEWithLogitsLoss()

# ---------------------------------------
# Boundary Loss
# ---------------------------------------

class BoundaryLoss(nn.Module):

    def __init__(self):
        super().__init__()

    def forward(self,pred,target):

        pred = torch.sigmoid(pred)

        pred_dx = torch.abs(
            pred[:,:,1:,:] -
            pred[:,:,:-1,:]
        )

        pred_dy = torch.abs(
            pred[:,:,:,1:] -
            pred[:,:,:,:-1]
        )

        target_dx = torch.abs(
            target[:,:,1:,:] -
            target[:,:,:-1,:]
        )

        target_dy = torch.abs(
            target[:,:,:,1:] -
            target[:,:,:,:-1]
        )

        loss_x = F.l1_loss(
            pred_dx,
            target_dx
        )

        loss_y = F.l1_loss(
            pred_dy,
            target_dy
        )

        return loss_x + loss_y


boundary_loss = BoundaryLoss()

# ---------------------------------------
# Combined Loss
# ---------------------------------------

def total_loss(pred,target):

    d = dice_loss(pred,target)

    b = bce_loss(pred,target)

    bd = boundary_loss(
        pred,
        target
    )

    return (

        0.5*d +

        0.3*b +

        0.2*bd
    )
```

```python
import os
import cv2
import random
import numpy as np

from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import Dataset
from torch.utils.data import DataLoader

import albumentations as A
from albumentations.pytorch import ToTensorV2

import segmentation_models_pytorch as smp

# =====================================================
# CONFIG
# =====================================================

IMAGE_DIR = r"D:\farmland boundary\filtered_patches\images"

MASK_DIR = r"D:\farmland boundary\filtered_patches\boundary_masks"

MODEL_PATH = "boundary_unetpp_b3_v2.pth"

PATCH_SIZE = 768

BATCH_SIZE = 2

EPOCHS = 60

LR = 1e-4

DEVICE = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)

print("DEVICE:", DEVICE)

# =====================================================
# DATASET
# =====================================================

files = sorted(os.listdir(IMAGE_DIR))

train_files, val_files = train_test_split(
    files,
    test_size=0.2,
    random_state=42
)

# =====================================================
# AUGMENTATION
# =====================================================

train_tf = A.Compose([

    A.HorizontalFlip(p=0.5),

    A.VerticalFlip(p=0.5),

    A.RandomRotate90(p=0.5),

    A.RandomBrightnessContrast(p=0.5),

    A.GaussianBlur(p=0.2),

    ToTensorV2()

])

val_tf = A.Compose([

    ToTensorV2()

])

# =====================================================
# DATASET
# =====================================================

class BoundaryDataset(Dataset):

    def __init__(
        self,
        file_list,
        transform
    ):

        self.files = file_list

        self.transform = transform

    def __len__(self):

        return len(self.files)

    def __getitem__(self,idx):

        file = self.files[idx]

        image = cv2.imread(
            os.path.join(
                IMAGE_DIR,
                file
            )
        )

        image = cv2.cvtColor(
            image,
            cv2.COLOR_BGR2RGB
        )

        mask = cv2.imread(
            os.path.join(
                MASK_DIR,
                file
            ),
            0
        )

        mask = (
            mask > 127
        ).astype(np.float32)

        transformed = self.transform(

            image=image,

            mask=mask
        )

        image = (
            transformed["image"]
            .float()
            /255.0
        )

        mask = (
            transformed["mask"]
            .unsqueeze(0)
            .float()
        )

        return image, mask

# =====================================================
# DATALOADER
# =====================================================

train_ds = BoundaryDataset(
    train_files,
    train_tf
)

val_ds = BoundaryDataset(
    val_files,
    val_tf
)

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

# =====================================================
# MODEL
# =====================================================

model = smp.UnetPlusPlus(

    encoder_name="efficientnet-b3",

    encoder_weights="imagenet",

    in_channels=3,

    classes=1

)

model = model.to(DEVICE)

# =====================================================
# LOSS
# =====================================================

dice_loss = smp.losses.DiceLoss(
    mode="binary"
)

bce_loss = nn.BCEWithLogitsLoss()

# =====================================================
# BOUNDARY LOSS
# =====================================================

class BoundaryLoss(nn.Module):

    def __init__(self):

        super().__init__()

    def forward(
        self,
        pred,
        target
    ):

        pred = torch.sigmoid(pred)

        pred_dx = torch.abs(
            pred[:,:,1:,:]
            -
            pred[:,:,:-1,:]
        )

        pred_dy = torch.abs(
            pred[:,:,:,1:]
            -
            pred[:,:,:,:-1]
        )

        target_dx = torch.abs(
            target[:,:,1:,:]
            -
            target[:,:,:-1,:]
        )

        target_dy = torch.abs(
            target[:,:,:,1:]
            -
            target[:,:,:,:-1]
        )

        loss_x = F.l1_loss(
            pred_dx,
            target_dx
        )

        loss_y = F.l1_loss(
            pred_dy,
            target_dy
        )

        return loss_x + loss_y

boundary_loss = BoundaryLoss()

def total_loss(
    pred,
    target
):

    d = dice_loss(
        pred,
        target
    )

    b = bce_loss(
        pred,
        target
    )

    bd = boundary_loss(
        pred,
        target
    )

    return (

        0.5*d

        +

        0.3*b

        +

        0.2*bd
    )

# =====================================================
# IOU
# =====================================================

def iou_score(
    pred,
    target
):

    pred = (
        torch.sigmoid(pred)
        > 0.5
    )

    inter = (
        pred & target.bool()
    ).float().sum()

    union = (
        pred | target.bool()
    ).float().sum()

    return (
        inter+1e-6
    ) / (
        union+1e-6
    )

# =====================================================
# OPTIMIZER
# =====================================================

optimizer = torch.optim.AdamW(

    model.parameters(),

    lr=LR,

    weight_decay=1e-4
)

scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(

    optimizer,

    T_max=EPOCHS
)

scaler = torch.cuda.amp.GradScaler()

# =====================================================
# TRAIN
# =====================================================

best_iou = 0

for epoch in range(EPOCHS):

    # -------------------------
    # TRAIN
    # -------------------------

    model.train()

    train_loss = 0

    for images,masks in train_loader:

        images = images.to(DEVICE)

        masks = masks.to(DEVICE)

        optimizer.zero_grad()

        with torch.cuda.amp.autocast():

            outputs = model(images)

            loss = total_loss(
                outputs,
                masks
            )

        scaler.scale(loss).backward()

        scaler.step(
            optimizer
        )

        scaler.update()

        train_loss += loss.item()

    train_loss /= len(train_loader)

    # -------------------------
    # VALIDATION
    # -------------------------

    model.eval()

    val_loss = 0

    val_iou = 0

    with torch.no_grad():

        for images,masks in val_loader:

            images = images.to(DEVICE)

            masks = masks.to(DEVICE)

            outputs = model(images)

            loss = total_loss(
                outputs,
                masks
            )

            val_loss += loss.item()

            val_iou += iou_score(
                outputs,
                masks
            ).item()

    val_loss /= len(val_loader)

    val_iou /= len(val_loader)

    scheduler.step()

    print(
        f"Epoch {epoch+1}/{EPOCHS}"
        f" | Train {train_loss:.4f}"
        f" | Val {val_loss:.4f}"
        f" | IoU {val_iou:.4f}"
    )

    if val_iou > best_iou:

        best_iou = val_iou

        torch.save(
            model.state_dict(),
            MODEL_PATH
        )

        print(
            "BEST MODEL SAVED"
        )

print("TRAINING COMPLETE")
print("BEST IOU:", best_iou)
```
